#Funções

```
# Função utilizada para gerar os dados truth para simulações 
# devemos passar para esta função:
# N_amostras= número de amostras à serem geradas
# p_luminos= porcentagem de luminosidade
# retorna os pulsos sorteados e um vetor com a posição de cada pulso

def simdistuni(N_amostras,p_luminos,amplitude_max):
  """ Função utilizada para gerar os dados truth para simulações
  Returns :  
  S_truth :
    An array with the truth data pulses for the simulations.
  posicao : 
    An array with the position for each pulse.
  """

  # faço o calculo da porcentagem da luminosidade
  p_luminos = p_luminos/100

  # crio um vetor de zeros para receber os impulsos de amplitudes sorteadas 
  S_truth   = np.zeros(N_amostras)

  # sorteio a posição das amostras "OBS:REQUER UM MINIMO DE UM PULSO por conta do rand "
  # >>> existe random.permutation mas so aceita 1 parametro... ai nao entrava o numero de amostras
  posicao   = np.random.choice(N_amostras, math.ceil(len(S_truth)*p_luminos)) 
  
  # sorteio a amplitude na distribuição uniforme
  amp_aleatoria = amplitude_max*np.random.rand(len(posicao))
  
  for i in range (0,len(posicao)):
    S_truth[posicao[i]] = amp_aleatoria[i]

  return S_truth, posicao
```

```
def simdistexp(N_amostras,p_luminos,media_exp):
  """
  Função utilizada para gerar os dados truth para simulações
  """

  # faço o calculo da porcentagem da luminosidade
  p_luminos = p_luminos/100

  # crio um vetor de zeros para receber os impulsos de amplitudes sorteadas 
  S_truth   = np.zeros(N_amostras)

  # sorteio a posição das amostras "OBS:REQUER UM MINIMO DE UM PULSO por conta do rand "
  # existe random.permutation mas so aceita 1 parametro... ai nao entrava o numero de amostras
  posicao   = np.random.choice(N_amostras, math.ceil(len(S_truth)*p_luminos)) 
  
  # sorteio a amplitude na distribuição uniforme
  amp_aleatoria = np.random.exponential(media_exp,len(posicao))

  for i in range (0,len(posicao)):
    S_truth[posicao[i]] = amp_aleatoria[i]

  return S_truth,posicao
```

```
def conv_tilecal_pulse(S_truth,desvio,posicao1):
  """
  Função faz a convolução de 2 sinais caso o desvio seja 0 e simula um desvio de fase e convolui caso o desvio seja 1
  """

  matriz_sorteio = pulsos_fases
  N_A = len(matriz_sorteio)
  c = len(matriz_sorteio[0])

  #Realizo o sorteio do desvio de fase de -5 a +5 ns
  fase_posicao = np.random.randint([1,c],len(posicao1))
  centro = 0.5 + c/2
  centro = int(centro)
  a = (N_A-1)/2
  a = int(a)

  #criando o sinal1 com o comprimento de termos da convolução
  sinal1 = np.zeros(len(S_truth)+N_A-1)

  #colocando zeros nas bordas do sinal truth para entrar na "convolução" 
  #S_truth1 = [np.zeros(a), S_truth, np.zeros(a)]
  S_truth1 = []
  S_truth1 = [0] * a + S_truth.tolist() + [0] * a
  S_truth1 = np.array(S_truth1)

  jj = 0

  for i in range(a, len(S_truth1)-a):
    aux = np.zeros(N_A)
    if S_truth1[i]!=0:
      if desvio == 1:   
        for j in range(0,N_A-1):
          #algoritmo que convolui sorteando o desvio de fase                                               
          aux[j] = S_truth1[i]*matriz_sorteio[j][fase_posicao[jj]]
          # caso queira apenas a convolução basta mudar  o vetor matrix para o vetor a ser convoluido 
      else:
        for j in range(0,N_A-1):
          aux[j] = S_truth1[i]*matriz_sorteio[j][centro]
          # caso o desvio de fase nao seja simulado basta escolher aqui a coluna da matriz sorteio central para convolução neste caso 6
        jj=jj+1
    #caso queira convoluir com apenas uma fase da matrix substituir fase posição pelo numero da coluna da fase
    sinal1[i-a:i+a+1]=aux+sinal1[i-a:i+a+1] 

  return sinal1,S_truth1
```
